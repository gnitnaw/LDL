## Import modules

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LightSource
from numpy.random import rand
from numpy.linalg import inv
from math import pi
print(np.__version__)
```

## Set random seed

```python
np.random.seed(1357)
```

## Reproduce Figure A-7

```python
a = 0.5
b = 1
rr = 30
def f(x1, x2, theta):
    return np.power((x1*np.cos(theta)-x2*np.sin(theta))/a,2)+np.power((x1*np.sin(theta)+x2*np.cos(theta))/b,2)-1

x1 = np.array([-1,1])
x2 = np.array([-1,1])
th = np.linspace(pi/4-0.1, pi/4+0.1, rr)

fig = plt.figure()
ax = plt.axes(projection='3d')
X1, X2 = np.meshgrid(x1, x2)
for i in range(rr):
    Z = f(X1, X2, th[i])
    ax.plot_surface(X1, X2, Z, color='yellow');
plt.xlim(-3,3)
plt.ylim(-3,3)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('z');
ax.view_init(30, 55)
# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
fig = plt.figure()

#ax = fig.gca(projection='3d')
#ls = LightSource(azdeg=180, altdeg=65)
#cols = ls.shade(z, plt.cm.winter)
#surf = ax.plot_surface(x, y, z, rstride=1, cstride=1, linewidth=0,
#                       antialiased=False, facecolors=cols)
plt.show()
# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
```
