## Import modules

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LightSource
from numpy.random import rand
from numpy.linalg import inv
from math import pi
print(np.__version__)
```

## Set random seed

```python
np.random.seed(1357)
```

## Reproduce Figure A-7

```python
a = 0.5
b = 1.414
rr = 100
def f(x1, x2, theta):
    return np.power((x1*np.cos(theta)-x2*np.sin(theta))/a,2)+np.power((x1*np.sin(theta)+x2*np.cos(theta))/b,2)-1

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

x1 = np.linspace(-1,1,rr)
x2 = np.linspace(-1,1,rr)
th = pi/0.4

fig = plt.figure()
ax = plt.axes(projection='3d')
X1, X2 = np.meshgrid(x1, x2)
for i in range(rr):
    Z = f(X1, X2, th)
    ax.plot_surface(X1, X2, Z, color='yellow');
plt.xlim(-3,3)
plt.ylim(-3,3)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('z');
ax.view_init(30, 70)
# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
fig = plt.figure()
plt.show()
# %%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%%
```
